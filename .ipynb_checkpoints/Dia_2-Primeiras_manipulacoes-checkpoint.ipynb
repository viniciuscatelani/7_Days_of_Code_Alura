```python
# Importando bibliotecas necessárias
import os
import pandas as pd
import numpy as np
```

```python
# Diertório com os dados completos de empréstimo

dir_data = 'C:/Users/vinic/OneDrive/Documentos/GitHub/7_Days_of_Code_Alura/Data/'
```

```python
# Leitura dos dados completos de empréstimo

emprestimos_completo = pd.read_csv(os.path.join(dir_data, 'emprestimos_completo.csv'))
emprestimos_completo
```

# Criação e exclusão das colunas necessárias

```python
# Transformando coluna de localização em numérico, já que só contém números
emprestimos_completo['localizacao'] = emprestimos_completo['localizacao'].astype(int)
```

```python
# Definindo as condições e os valores correspondentes
conditions = [
    (emprestimos_completo['localizacao'] < 100),
    (emprestimos_completo['localizacao'] >= 100) & (emprestimos_completo['localizacao'] < 200),
    (emprestimos_completo['localizacao'] >= 200) & (emprestimos_completo['localizacao'] < 300),
    (emprestimos_completo['localizacao'] >= 300) & (emprestimos_completo['localizacao'] < 400),
    (emprestimos_completo['localizacao'] >= 400) & (emprestimos_completo['localizacao'] < 500),
    (emprestimos_completo['localizacao'] >= 500) & (emprestimos_completo['localizacao'] < 600),
    (emprestimos_completo['localizacao'] >= 600) & (emprestimos_completo['localizacao'] < 700),
    (emprestimos_completo['localizacao'] >= 700) & (emprestimos_completo['localizacao'] < 800),
    (emprestimos_completo['localizacao'] >= 800) & (emprestimos_completo['localizacao'] < 900),
    (emprestimos_completo['localizacao'] >= 900) & (emprestimos_completo['localizacao'] < 1000)
]

values = [
    'Generalidades, Ciência e Conhecimento',
    'Filosofia e Psicologia',
    'Religião',
    'Ciências sociais',
    'Classe vaga',
    'Matemática e ciências naturais',
    'Ciências aplicadas',
    'Belas artes',
    'Linguagem. Língua. Linguística',
    'Geografia. Biografia. História'
]

# Aplicando as condições ao DataFrame
emprestimos_completo['classe_cdu'] = np.select(conditions, values)
```

```python
emprestimos_completo
```

```python
# Excluindo coluna 'registro_sistema'

emprestimos_completo = emprestimos_completo.drop('registro_sistema', axis=1)
emprestimos_completo
```

```python
# Modificando coluna 'matriucla_ou_siape' para string

emprestimos_completo['matricula_ou_siape'] = emprestimos_completo['matricula_ou_siape'].astype(str)
emprestimos_completo
```

```python
emprestimos_completo.to_csv(os.path.join(dir_data, 'emprestimos_completo.csv'))
```
