# Função para visuilzação de tabela com as informações desejadas

```python
# Definindo função para gerar as tabelas
def tab_frequencia(variavel):
    
    grouped = (emprestimos_completo
               .groupby(variavel)
               .agg({'id_emprestimo': 'count'})
              )
    grouped.rename(columns={'id_emprestimo': 'quantidade'}, inplace=True)
    grouped['percentual'] = round((grouped['quantidade']/grouped['quantidade'].sum())*100, 2)
    grouped = grouped.sort_values(by='percentual', ascending=False)
    
    return grouped
```

```python
# Gerando tabela de frequência para tipo de vínculo

tipo_vinculo = tab_frequencia('tipo_vinculo_usuario')
tipo_vinculo
```

É possível perceber que os que mais realizam empréstimos de livros são os alunos de graduação, seguidos dos alunos de pós-graduação.

A grande diferença na porcentagem entre ambos pode ser devido a fatores como:
   - A diferença de quantidade de alunos de graduação em comparação com os de pós-graduação.
   - O tipo de conteúdo que os alunos de pós-graduação mais utilizam devido ao seu objetivo.

```python
# Gerando tabela de frequência para coleção

colecao = tab_frequencia('colecao')
colecao
```

Ao analisar a tabela, percebe-se que quase todos os livros emprestadpos fazem parte do acervo circulante.
