```python
# Importando bibliotecas necessárias

import os
import pandas as pd
```

```python
# Diertório com os dados completos de empréstimo

dir_data = 'C:/Users/vinic/OneDrive/Documentos/GitHub/7_Days_of_Code_Alura/Python Pandas/Data/'
```

```python
# Lendo o arquivo com os dados de empréstimo

emprestimos_completo = pd.read_csv(os.path.join(dir_data, 'emprestimos_completo.csv'))
emprestimos_completo = emprestimos_completo.drop(['Unnamed: 0.1', 'Unnamed: 0'], axis=1)
emprestimos_completo
```

```python
# Seleção dos empréstimo dos alunos de pós-graduação

pos_grad = emprestimos_completo.query('tipo_vinculo_usuario == "ALUNO DE PÓS-GRADUAÇÃO"').reset_index(drop=True)
pos_grad
```

```python
# Lendo os arquivos com os dados de matricula dos alunos depois de 2010

matricula_pos_2010 = pd.read_excel(os.path.join(dir_data, 'matricula_alunos.xlsx'),
                          sheet_name='Após 2010',
                          header=1)

matricula_pos_2010['Matricula '] = matricula_pos_2010['Matricula '].astype(str)
matricula_pos_2010.rename(columns={'Matricula ': 'matricula_ou_siape',
                                  'Tipo de vinculo': 'tipo_vinculo_usuario',
                                  'Cursos': 'curso'}, inplace=True)

matricula_pos_2010
```

```python
# Lendo os arquivos com os dados de cadastro dos alunos
cadastro_json = pd.read_json(os.path.join(dir_data, 'cadastro_alunos.json'))
cadastro = pd.read_json(cadastro_json.registros[1])
cadastro['matricula_ou_siape'] = cadastro['matricula_ou_siape'].astype(float).astype(str)
cadastro
```

```python
# Concatenação das planilhas do excel com o JSON

cadastro_geral = pd.concat([matricula_pos_2010, cadastro])
cadastro_geral
```

```python
# Extraindo informaçõe dos empréstimos a partir de 2017 até 2019

pos_grad['data_emprestimo'] = pd.to_datetime(pos_grad['data_emprestimo']).dt.year
emprestimos_17_a_19 = pos_grad[pos_grad['data_emprestimo'].between(2017, 2019)]


emprestimos_17_a_19 = emprestimos_17_a_19[['matricula_ou_siape', 'data_emprestimo']].reset_index(drop=True)

emprestimos_17_a_19['matricula_ou_siape'] = emprestimos_17_a_19['matricula_ou_siape'].astype(str)
emprestimos_17_a_19
```

```python
# Ajuste do tipo da coluna

cadastro['matricula_ou_siape'] = cadastro['matricula_ou_siape'].astype(str)
```

```python
# Mesclagem dos dados

merged = emprestimos_17_a_19.merge(cadastro)
merged
```

```python
# Geração da tabela com as informações necessárias

data = (merged
       .groupby(['data_emprestimo', 'curso'], as_index=False)
       .agg({'matricula_ou_siape': 'count'})
       ).sort_values(by=['curso', 'data_emprestimo'])

data.rename(columns={'data_emprestimo': 'ano',
                    'matricula_ou_siape': 'quantidade'}, inplace=True)

data = data.reset_index(drop=True)
data['ano'] = data['ano'].astype(int)
data
```

```python
# Geração de tabela com as caracteristicas necessarias

emprestimos_pivot = data.pivot_table(index='curso',
                                    columns='ano',
                                    values='quantidade',
                                     aggfunc='sum',
                                    )

emprestimos_pivot
```

```python
# Leitura do arquivo referent ao ano de 2022

dados_2022 = pd.read_table(os.path.join(dir_data, 'previsao.txt'), sep=' ')
dados_2022
```

```python
# Formatação das colunas e das informações

dados_2022.rename(columns={'previsao_2022': 'quantidade'}, inplace=True)
dados_2022['ano'] = [2022]*dados_2022.shape[0]

dados_2022 = dados_2022[['ano', 'curso', 'quantidade']]
dados_2022 = dados_2022.set_index('curso')
dados_2022.index = emprestimos_pivot.index
# dados_2022['ano'] = dados_2022['ano'].astype(int)
dados_2022
```

```python
# Geração de tabela com as caracteristicas necessarias

emprestimos_pivot['2022'] = dados_2022.iloc[:, [1]]
emprestimos_pivot
```

```python
# Garantir que as colunas sejam tratadas como strings
emprestimos_pivot.columns = emprestimos_pivot.columns.astype(str)

# Calcula a diferença percentual entre os anos especificados
emprestimos_pivot['diff_2017_2018'] = ((emprestimos_pivot['2018'] - emprestimos_pivot['2017']) / emprestimos_pivot['2017']) * 100
emprestimos_pivot['diff_2018_2019'] = ((emprestimos_pivot['2019'] - emprestimos_pivot['2018']) / emprestimos_pivot['2018']) * 100
emprestimos_pivot['diff_2019_2022'] = ((emprestimos_pivot['2022'] - emprestimos_pivot['2019']) / emprestimos_pivot['2019']) * 100

# Mantém apenas as colunas de interesse
diferencas_percentuais = emprestimos_pivot[['diff_2017_2018', 'diff_2018_2019', 'diff_2019_2022']]

diferencas_percentuais = diferencas_percentuais.reset_index()
diferencas_percentuais
```

```python
# Formatando as colunas e os dados

diferencas_percentuais.columns = ['Curso', '2018', '2019', '2022']
diferencas_percentuais.Curso = diferencas_percentuais.Curso.str.capitalize()
diferencas_percentuais
```

# Gerando tabela em HTML

```python
# Definindo os parâmetros da estilização CSS
styles = [
    dict(selector='th', props = [
        ('font-size', '1.4rem'),
        ('text-align', 'center'),
        ('font-weight', 'bold'),
        ('color', 'whitesmoke'),
        ('background-color', '#001692'),
        ('border-radius', '0.25rem'),
        ('box-shadow', '0.01rem gray')
    ]
        ),

    dict(selector='td', props = [
        ('font-size', '1rem'),
        ('padding', '0.5rem'),
        ('text-align', 'left'),
        ('font-weight', 'bold'),
        ('border-bottom', '0.1rem solid lightgray')
    ]
        )
]
```

```python
diferencas_percentuais.style.text_gradient(cmap='RdYlGn',low=1, axis=1,vmax=0.1,vmin=0)\
                              .format('{:.2f} %',subset=['2018','2019','2022'])\
                              .hide_index()\
                              .set_table_styles(styles)\
                              .to_html('percentuais.html',doctype_html =True,
                                       table_attributes='ALIGN=LEFT WIDTH=50% CELLSPACING = 5')
```
