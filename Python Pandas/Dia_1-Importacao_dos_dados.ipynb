```python
# Importação das bibliotecas necessárias
import pandas as pd
import os
```

# Leitura dos arquivos 

```python
# Leitura dos arquivos de empréstimo em uma única tabela

# Definindo diretório dos dados
dir_data = 'C:/Users/vinic/OneDrive/Documentos/GitHub/7_Days_of_Code_Alura/Python Pandas/Data/'

# Leitura dos arquivos
tabelas = []

for file in os.listdir(dir_data):
    if file.endswith('.csv'):
        data = pd.read_csv(os.path.join(dir_data, file))
        tabelas.append(data)
        
emprestimos = pd.concat(tabelas)
```

```python
# Visualização do dataframe final para melhor entendimento inicial dos dados
emprestimos['data_emprestimo'] = pd.to_datetime(emprestimos['data_emprestimo'])
emprestimos['data_devolucao'] = pd.to_datetime(emprestimos['data_devolucao'])

print(emprestimos.shape)
emprestimos.head()
```

Já nas primeiras linhas é possível ver que há dados nulos e linhas duplicadas, por exemplo, a quinta linha é uma duplicação da primeira linha, portanto em etapas seguintes será necessário eliminar as linhas duplicadas.

Sobre os dados nulos, é necessário, principalmente se a intenção é excluir linhas com dados nulos, se a quantidade de dados nulos não é grande, pois essa quantidade pode ser um indicador importante para futuras análises.

```python
# Leitura do arquivo dos dados do acervo

acervo = pd.read_parquet(os.path.join(dir_data, 'dados_exemplares.parquet'))
```

```python
# Visualização inicial dos dados do acervo
acervo
```

# Identificação e limpeza de dados nulos e duplicados

```python
# Visualização dos primeiros dados 

emprestimos.head(10)
```

```python
# Identificaão de quantas linhas duplicadas existem no dataframe
emprestimos[emprestimos.duplicated()].shape
```

Existem poucas linhas duplicadas no dataframe, portanto é possível excluí-las sem prejuízo ao dataframe final para análises.

Caso o número de linhas fosse grande demais, isso poderia representar um possível erro na coleta ou até registros dos dados, não é possível dizer qual dos dois seria, ou se seria algum outro possível problema, porém teria de ser investigado para entender mais a fundo o porquê de tantas duplicadas.

```python
# Filtrando dataframe para exclusão das linhas duplicadas
emprestimos = emprestimos.drop_duplicates()
print(emprestimos.shape)
emprestimos.head(10)
```

```python
# Visualização sobre informações mais detalhadas dos dados
emprestimos.info(show_counts=True)
```

É possível perceber que para as colunas **data_renovacao** e **data_devolucao** contêm valores nulos para algumas das linhas, porém, para a primeira, mais da metade dos dados são nulos, portanto não é possível excluir as linhas que possuem dados nulos, pois isso irá comprometer futuras análises.

Analisando as quantidades de valores nulos, é possível inferir que:

    - A maioria das pessoas não renovam os empréstimos de livros;
    - Uma quantidade muito pequena de pessoas não devolvem os livros emprestados;

```python
# Identificação de linhas duplicadas no arquivo do acervo

acervo[acervo.duplicated()].shape
```

Não existem linhas duplicadas no arquivo do acervo, então vamos seguir para analisar os dados nulos

```python
# Visualização das informações do acervo
acervo.info()
```

Não existem valores nulos para nenhum dado nos arquivos do acervo, o que é uma notícia boa, não será necessário pensar em excluir qualquer dado do dataframe

# Mesclagem das tabelas

Ambos os dataframes se relacionam a partir da coluna **codigo_barras**, portanto, para manter todas as informações necessárias em uma única tabela, é necessários mesclá-las.

**Detalhe muito importante**: é sempre necessário garantir que, em ambos os dataframes, o tipo das colunas que serão a referência para operação de merge seja o mesmo, do contrário será retornado um erro.

```python
# Mesclagem das tabelas

merged = emprestimos.merge(acervo, on='codigo_barras')
merged
```

```python
merged.to_csv(os.path.join(dir_data, 'emprestimos_completo.csv'))
```
