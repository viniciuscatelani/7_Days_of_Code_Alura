```python
# Importando bibliotecas necessárias

import os
import pandas as pd
import numpy as np
```

```python
# Diertório com os dados completos de empréstimo

dir_data = 'C:/Users/vinic/OneDrive/Documentos/GitHub/7_Days_of_Code_Alura/Data/'
```

```python
# Lendo o arquivo com os dados de empréstimo

emprestimos_completo = pd.read_csv(os.path.join(dir_data, 'emprestimos_completo.csv'))
emprestimos_completo = emprestimos_completo.drop(['Unnamed: 0.1', 'Unnamed: 0'], axis=1)
emprestimos_completo
```

```python
# Lendo os arquivos com os dados de matricula dos alunos até 2010

matricula_ate_2010 = pd.read_excel(os.path.join(dir_data, 'matricula_alunos.xlsx'),
                          sheet_name='Até 2010',
                          header=1)

matricula_ate_2010['Número da matrícula'] = matricula_ate_2010['Número da matrícula'].astype(str)
matricula_ate_2010.rename(columns={'Número da matrícula': 'matricula_ou_siape',
                                  'Tipo de vinculo': 'tipo_vinculo_usuario',
                                  'Curso': 'curso'}, inplace=True)
matricula_ate_2010
```

```python
# Lendo os arquivos com os dados de matricula dos alunos depois de 2010

matricula_pos_2010 = pd.read_excel(os.path.join(dir_data, 'matricula_alunos.xlsx'),
                          sheet_name='Após 2010',
                          header=1)

matricula_pos_2010['Matricula '] = matricula_pos_2010['Matricula '].astype(str)
matricula_pos_2010.rename(columns={'Matricula ': 'matricula_ou_siape',
                                  'Tipo de vinculo': 'tipo_vinculo_usuario',
                                  'Cursos': 'curso'}, inplace=True)

matricula_pos_2010
```

```python
# Lendo os arquivos com os dados de cadastro dos alunos
cadastro_json = pd.read_json(os.path.join(dir_data, 'cadastro_alunos.json'))
cadastro = pd.read_json(cadastro_json.registros[0])
cadastro['matricula_ou_siape'] = cadastro['matricula_ou_siape'].astype(float).astype(str)
cadastro
```

```python
# concatenação das duas planilhas do arquivo excel

cadastro_excel = pd.concat([matricula_ate_2010, matricula_pos_2010],ignore_index=True)
cadastro_excel['matricula_ou_siape'] = cadastro_excel['matricula_ou_siape'].astype(str) 
cadastro_excel
```

```python
# Concatenação das planilhas do excel com o JSON

cadastro_geral = pd.concat([cadastro_excel, cadastro])
cadastro_geral
```

```python
# Extraindo informaçõe dos empréstimos a partir de 2015

emprestimos_2015 = emprestimos_completo[emprestimos_completo['data_emprestimo'] > '2015-01-01']
# emprestimos_2015['matricula_ou_siape'] = emprestimos_2015['matricula_ou_siape'].astype(str)
emprestimos_2015 = emprestimos_2015[emprestimos_2015['tipo_vinculo_usuario'] == 'ALUNO DE GRADUAÇÃO'][['matricula_ou_siape', 
                                                                                                'data_emprestimo']].reset_index(drop=True)
emprestimos_2015
```

```python
# verificando informações das colunas da tabela de emprestimos

emprestimos_2015.info()
```

```python
# retirada dos valores nulos
emprestimos_2015 = emprestimos_2015.dropna()
```

```python
# Seleção dos alunos com emprestimos dos cursos desejados

cursos = ['BIBLIOTECONOMIA', 'CIÊNCIAS SOCIAIS', 'COMUNICAÇÃO SOCIAL',
         'DIREITO', 'FILOSOFIA', 'PEDAGOGIA']

cadastro_filtered = cadastro_geral[cadastro_geral['curso'].isin(cursos)]
cadastro_filtered
```

# Mesclagem dos dados de cadastro com os dados de empréstimo

```python
# Ajuste do tipo de coluna

emprestimos_2015['matricula_ou_siape'] = emprestimos_2015['matricula_ou_siape'].astype(str)

cadastro_filtered['matricula_ou_siape'] = cadastro_filtered['matricula_ou_siape'].astype(str)
```

```python
# Mesclagem dos dados

merged = emprestimos_2015.merge(cadastro_filtered)
merged
```

# Geração da tabela de quantidade de empréstimos por curso por ano

```python
# extração da informação de ano da coluna de data de empréstimo

merged['data_emprestimo'] = pd.to_datetime(merged['data_emprestimo'])
merged['data_emprestimo'] = merged['data_emprestimo'].dt.year
merged
```

```python
# Geração da tabela com as informações necessárias

data = (merged
       .groupby(['data_emprestimo', 'curso'], as_index=False)
       .agg({'matricula_ou_siape': 'count'})
       ).sort_values(by=['curso', 'data_emprestimo'])

data.rename(columns={'data_emprestimo': 'ano',
                    'matricula_ou_siape': 'quantidade'}, inplace=True)

data
```

```python
# Geração de tabela com as caracteristicas necessarias

emprestimos_pivot = data.pivot_table(index='curso',
                                    columns='ano',
                                    values='quantidade',
                                     aggfunc='sum',
                                    margins=True,
                                    margins_name='Total')

emprestimos_pivot
```
